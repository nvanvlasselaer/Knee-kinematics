**confidence intervals, using bootstrapping**

```python
import numpy as np
import matplotlib.pyplot as plt
from skfda import FDataGrid
from skfda.preprocessing.dim_reduction import FPCA
from sklearn.cluster import KMeans

# Function for bootstrapping
def bootstrap_samples(data, n_bootstrap):
    n_samples = data.shape[0]
    bootstrap_indices = np.random.choice(np.arange(n_samples), size=(n_bootstrap, n_samples), replace=True)
    bootstrap_samples = [data[indices] for indices in bootstrap_indices]
    return bootstrap_samples

# Prepare functional data
X = []  # List to store spline regression curves for each subject
subject_names = []  # List to store subject names
for subject_data, subject_name in subjects_zero_L_OKC:
    X.append(subject_data.values)
    subject_names.append(subject_name)

X = np.array(X)  # Convert to numpy array

# Create FDataGrid object
fd_grid = FDataGrid(data_matrix=X)

# Apply Functional Principal Component Analysis (FPCA)
n_components = 20  # Number of principal components
fpca = FPCA(n_components=n_components)
fd_grid_fpca = fpca.fit_transform(fd_grid)

# Bootstrap resampling
n_bootstrap = 100  # Number of bootstrap samples
bootstrap_fpca_samples = bootstrap_samples(fd_grid_fpca, n_bootstrap)

# Perform clustering on the FPCA scores of bootstrap samples
n_clusters = 4  # Number of clusters
cluster_assignments = np.zeros((n_bootstrap, fd_grid_fpca.shape[0]), dtype=np.int32)
for i, sample in enumerate(bootstrap_fpca_samples):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10)
    cluster_assignments[i] = kmeans.fit_predict(sample)

# Calculate confidence intervals
confidence_intervals = []
for cluster_id in range(n_clusters):
    cluster_indices = np.where(clusters == cluster_id)[0]
    cluster_counts = np.sum(cluster_assignments == cluster_id, axis=0)
    lower_bound = np.percentile(cluster_counts, 2.5)
    upper_bound = np.percentile(cluster_counts, 97.5)
    confidence_intervals.append((lower_bound, upper_bound))

# Print confidence intervals
for i, (lower_bound, upper_bound) in enumerate(confidence_intervals):
    print(f"Cluster {i}: Confidence Interval (2.5% - 97.5%): [{lower_bound}, {upper_bound}]")

# Visualize clusters
plt.figure(figsize=(9, 6))
for i in range(n_clusters):
    cluster_indices = np.where(clusters == i)[0]
    for idx in cluster_indices:
        plt.plot(fd_grid[idx].grid_points[0], fd_grid[idx].data_matrix[0], 
                 label=f'{i}: {subject_names[idx]}', color=colors[i])

plt.xlabel('Knee Flexion')
plt.ylabel('Rotation')
plt.title('Functional Data Clustering with Confidence Intervals')
plt.axhline(y=0, color='black', linestyle='--')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()

```

The confidence intervals indicate the range of variability in the number of knee rotation registrations assigned to each cluster,

Cluster 0: confidence interval [24.1, 38.45] suggests that, with 95% confidence, the true number of knee rotation registrations assigned to Cluster 0 falls within the range of approximately 24.1 to 38.45. Cluster 0 is expected to contain knee rotation registrations ranging from about 24 to 38 for each degree of knee flexion.

Cluster 1: Similarly, for Cluster 1, the confidence interval [21.1, 37.45] indicates that the number of knee rotation registrations assigned to this cluster is expected to vary between approximately 21.1 and 37.45 with 95% confidence. This range corresponds to the variability in knee rotation registrations observed for each degree of knee flexion within Cluster 1.

Cluster 2: confidence interval [14.1, 28.0] suggests a narrower range of knee rotation registrations compared to Clusters 0 and 1. With 95% confidence, Cluster 2 is expected to contain knee rotation registrations ranging from about 14 to 28 for each degree of knee flexion.

Cluster 3: confidence interval [9.55, 22.45] indicates a moderate variability in knee rotation registrations within this cluster. The range of knee rotation registrations expected in Cluster 3, with 95% confidence, is approximately 9.55 to 22.45 for each degree of knee flexion.

**Weighted Functional Data Clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from skfda import FDataGrid
from skfda.preprocessing.dim_reduction import FPCA
from sklearn.cluster import KMeans

# Define a function to generate weights
def generate_weights(n_points, weight_mode='linear', **kwargs):
    if weight_mode == 'linear':
        max_weight = kwargs.get('max_weight', 1.0)
        min_weight = kwargs.get('min_weight', 0.1)
        weights = np.linspace(max_weight, min_weight, n_points)
    elif weight_mode == 'exponential':
        decay_rate = kwargs.get('decay_rate', 0.1)
        weights = np.exp(-decay_rate * np.arange(n_points))
    elif weight_mode == 'custom':
        # Define custom weighting scheme here
        pass
    else:
        raise ValueError("Invalid weight_mode. Choose 'linear', 'exponential', or 'custom'.")
    return weights

# Prepare functional data
X = []  # List to store spline regression curves for each subject
subject_names = []  # List to store subject names
for subject_data, subject_name in subjects_zero_L_OKC:
    X.append(subject_data.values)
    subject_names.append(subject_name)

X = np.array(X)  # Convert to numpy array

# Generate weights
n_points = X.shape[1]
weights = generate_weights(n_points, weight_mode='linear', max_weight=1.0, min_weight=0.1)  # Adjust parameters as needed

# Apply weights to the functional data
X_weighted = X * weights

# Create FDataGrid object
fd_grid_weighted = FDataGrid(data_matrix=X_weighted)

# Apply Functional Principal Component Analysis (FPCA)
n_components = 6  # Number of principal components
fpca = FPCA(n_components=n_components)
fd_grid_fpca_weighted = fpca.fit_transform(fd_grid_weighted)

# Perform clustering on the FPCA scores
n_clusters = 5  # Number of clusters
kmeans = KMeans(n_clusters=n_clusters, n_init=10)  # Set n_init explicitly
clusters = kmeans.fit_predict(fd_grid_fpca_weighted)  # Access transformed data directly


# Print the clusters
for cluster_id in range(n_clusters):
    cluster_indices = np.where(clusters == cluster_id)[0]
    subject_names_in_cluster = [subject_names[idx] for idx in cluster_indices]
    print(f"Cluster {cluster_id}: {subject_names_in_cluster}")

colors = ['b', 'g', 'r', 'c', 'm', 'y']  # Add more colors if needed

# Visualize clusters
plt.figure(figsize=(9, 6))
for i in range(n_clusters):
    cluster_indices = np.where(clusters == i)[0]
    for idx in cluster_indices:
        plt.plot(fd_grid_weighted[idx].grid_points[0], fd_grid_weighted[idx].data_matrix[0], label=f'{i},: {subject_names[idx]}', color=colors[i])

plt.xlabel('Knee Flexion')
plt.ylabel('Rotation')
plt.title('Functional Data Clustering with Weighted Analysis')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
